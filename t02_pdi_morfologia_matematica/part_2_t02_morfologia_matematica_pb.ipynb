## Criando uma função para exibir as imagens em uma grid

```python
from matplotlib import pyplot as plt

    
def showMultipleImages(imgsArray, titlesArray, size, x, y, mincolor=0, maxcolor=255):
    if(x < 1 or y < 1):
        print("ERRO: X e Y não podem ser zero ou abaixo de zero!")
        return
    elif(x == 1 and y == 1):
        showSingleImage(imgsArray, titlesArray)
    elif(x == 1):
        fig, axis = plt.subplots(y, figsize = size)
        yId = 0
        for img in imgsArray:
            axis[yId].imshow(img, cmap='gray', vmin=mincolor, vmax=maxcolor)
            axis[yId].set_anchor('NW')
            axis[yId].set_title(titlesArray[yId], fontdict = {'fontsize': 18, 'fontweight': 'medium'}, pad = 10)

            yId += 1
    elif(y == 1):
        fig, axis = plt.subplots(1, x, figsize = size)
        fig.suptitle(titlesArray)
        xId = 0
        for img in imgsArray:
            axis[xId].imshow(img, cmap='gray', vmin=mincolor, vmax=maxcolor)
            axis[xId].set_anchor('NW')
            axis[xId].set_title(titlesArray[xId], fontdict = {'fontsize': 18, 'fontweight': 'medium'}, pad = 10)

            xId += 1
    else:
        fig, axis = plt.subplots(y, x, figsize = size)
        xId, yId, titleId = 0, 0, 0
        for img in imgsArray:
            axis[yId, xId].set_title(titlesArray[titleId], fontdict = {'fontsize': 18, 'fontweight': 'medium'}, pad = 10)
            axis[yId, xId].set_anchor('NW')
            axis[yId, xId].imshow(img, cmap='gray', vmin=mincolor, vmax=maxcolor)
            if(len(titlesArray[titleId]) == 0):
                axis[yId, xId].axis('off')

            titleId += 1
            xId += 1
            if xId == x:
                xId = 0
                yId += 1
    plt.show()
```

## Realizando as operações morfológicas:

### Para a realização das operações de ``Erosão``, ``Dilatação``, ``Abertura`` e ``Fechamento``, primeiramente se faz necessário definir o operador morfológico (elemento estruturante) também conhecido como ``kernel``

### Definindo o operador morfológico (quadrático):

```python
# Operador quadrático (1 x 1)
square_kernel_1 = cv2.getStructuringElement(cv2.MORPH_RECT, (1,1))

# Operador quadrático (2 x 2)
square_kernel_2 = cv2.getStructuringElement(cv2.MORPH_RECT, (2,2))

# Operador quadrático (3 x 3)
square_kernel_3 = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
```

### Definindo o operador morfológico (linear):

```python
# Operador linear (1 x 1)
linear_kernel_1 = np.ones((1, 1), np.uint8)

# Operador linear (1 x 2)
linear_kernel_2 = np.ones((1, 2), np.uint8)

# Operador linear (1 x 3)
linear_kernel_3 = np.ones((1, 3), np.uint8)
```

### Definindo o operador morfológico (circular):

```python
# Operador circular (1 x 1)
circle_kernel_1 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(1,1))

# Operador circular (2 x 2)
circle_kernel_2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(2,2))

# Operador circular (3 x 3)
circle_kernel_3 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
```
